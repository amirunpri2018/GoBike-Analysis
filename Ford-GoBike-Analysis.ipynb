#### Commuter Trips

The data shows that there are two types of riders:
* Subscribers, who have taken annual membership (most likely rush hour commuters)
* Customers, who use this service on an as-needed basis (day trip / visitor to the city)

The following query captures the details around these two categories of users and how their rides compare during the rush hour (early / morning commuters).  The result from this query is plotted to depict how the ridership considerably differs between subscribers and customers.


As a first step, let's get an overall understanding of how the trip counts have varied from August 2013 through August 2016.  The query below captures the details and the heatmap helps us visualize this key information.

```python
%%bash
bq query --use_legacy_sql=FALSE --format=csv '
with cte as
(
SELECT
  EXTRACT(YEAR FROM start_date) AS Year,
  subscriber_type as Subscriber_Type,
  case 
      when time(start_date) between "6:00:00" and "10:00:00" then "Morning_Commuter"
      when time(start_date) between "15:00:00" and "20:00:00" then "Evening_Commuter"
      else "Other"
  end as Commuter_Type
#  count(*) as Early_Morning_Commuters
FROM
  `bigquery-public-data.san_francisco.bikeshare_trips`
WHERE
  time(start_date) between "6:00:00" and "20:00:00"
)
select 
  *,
  count(*) as Total_Count
from 
  cte
GROUP BY  Year, Subscriber_Type, Commuter_Type
ORDER BY 2, 1 asc ' > Commuters_Counts.csv
```

### Analysis - How to Increase Ridership

The initial exploratory analysis on the bikeshare dataset has given some insights into the type of riders, their usage etc.  Now, let us focus on the analysis that will help us understand what recommendations can be made, that would help with increase in ridership and inturn an increased customer base.

The following queries, followed by visualizations will help provide answers to the question above.  To begin with, let us try to understand how count of trips vary by hour of the day.  For this analysis, let us focus on customers who ride during morning, afternoon and evening hours.  The query describes how we categorize the hours into the appropriate bucket to capture the interval.

```python
%%bash
bq query --use_legacy_sql=FALSE --format=csv '
with cte as 
(
SELECT
  start_station_Id,
  zip_code,
  subscriber_type,
  CASE
    WHEN duration_sec/60 > 30 THEN "More than 30 min"
    ELSE "Less than 30 min"
  END AS duration_bucket,
  CASE
    WHEN TIME(start_date) >= "6:00:00" AND TIME(start_date)  < "12:00:00" THEN "Morning"
    WHEN TIME(start_date) >= "12:00:00" AND TIME(start_date) < "15:00:00" Then "Afternoon"
    WHEN TIME(start_date) >= "15:00:00" AND TIME(start_date) < "20:00:00" Then "Evening"
    WHEN TIME(start_date) >= "20:00:00" AND TIME(start_date) < "23:00:00" Then "Night"
    WHEN TIME(start_date) >= "23:00:00" THEN "Late Night"
    WHEN TIME(start_date) < "6:00:00" THEN "Early Hours"
    END AS dayinterval,
  start_date,
  TIME(start_date),
  FORMAT_TIME("%H", TIME(start_date)) as Hour_Flag
FROM
  `bigquery-public-data.san_francisco.bikeshare_trips`
WHERE
  ((DATE(end_date) = DATE(start_date))
  OR
  (DATE(end_date) = DATE_ADD(DATE(start_date), INTERVAL 1 DAY)))
)
SELECT 
  subscriber_type as Subscriber_Type,
  Hour_Flag,
  dayinterval as Day_Interval,
  duration_bucket as Duration_Type,
  count(*) as Trip_Count
FROM
  cte
WHERE
  dayinterval in ("Morning", "Afternoon", "Evening")
GROUP BY Subscriber_Type, Hour_Flag, Day_Interval, Duration_Type
ORDER BY 2,1 ' > TripCount_By_Hour.csv
```

### Analysis - Usage by Station

As a next step, let's look at how the stations perform and where we see maximum number of trips being made etc.  This will also help us understand the most utilized vs. under utilized stations.


```python
%%bash
bq query --use_legacy_sql=FALSE --format=csv '
SELECT
  start_station_id,
  start_station_name as Start_Station,
  subscriber_type as Subscriber_Type,
  ROUND(AVG(duration_sec)/60, 2) as Average_Duration_Minutes,
  COUNT(*) as Total_Number_Trips,
  CASE
    WHEN TIME(start_date) >= "6:00:00" AND TIME(start_date)  < "12:00:00" THEN "Morning"
    WHEN TIME(start_date) >= "12:00:00" AND TIME(start_date) < "15:00:00" Then "Afternoon"
    WHEN TIME(start_date) >= "15:00:00" AND TIME(start_date) < "20:00:00" Then "Evening"
    WHEN TIME(start_date) >= "20:00:00" AND TIME(start_date) < "23:00:00" Then "Night"
    WHEN TIME(start_date) >= "23:00:00" THEN "Late Night"
    WHEN TIME(start_date) < "6:00:00" THEN "Early Hours"
  END AS Day_Interval
FROM
  `bigquery-public-data.san_francisco.bikeshare_trips`
GROUP BY start_station_id, Start_Station, Subscriber_Type, Day_Interval
#HAVING dayinterval in ('Morning', 'Evening')
ORDER BY 5 DESC, 4 DESC
LIMIT 100 ' > TripsByStation.csv
```

##### A new pricing category can be introduced to cater to riders who typically use it for an hour.  This would provide incentive to the customer who typically rides for 1 hour or less, although they pay for 2 hour package.

```python
%%bash
bq query --use_legacy_sql=FALSE --format=csv '
with cte as 
(
SELECT
  start_station_Id,
  zip_code,
  subscriber_type,
  duration_sec/60 as Duration_in_Minutes,
  CASE
    WHEN duration_sec/60 > 30 THEN "More than 30 Min "
    ELSE "Less than 30 min"
  END AS duration_bucket,
  CASE
    WHEN TIME(start_date) >= "6:00:00" AND TIME(start_date)  < "12:00:00" THEN "Morning"
    WHEN TIME(start_date) >= "12:00:00" AND TIME(start_date) < "15:00:00" Then "Afternoon"
    WHEN TIME(start_date) >= "15:00:00" AND TIME(start_date) < "20:00:00" Then "Evening"
    WHEN TIME(start_date) >= "20:00:00" AND TIME(start_date) < "23:00:00" Then "Night"
    WHEN TIME(start_date) >= "23:00:00" THEN "Late Night"
    WHEN TIME(start_date) < "6:00:00" THEN "Early Hours"
    END AS dayinterval,
  start_date,
  TIME(start_date),
  FORMAT_TIME("%H", TIME(start_date)) as Hour_Flag
FROM
  `bigquery-public-data.san_francisco.bikeshare_trips`
WHERE
  ((DATE(end_date) = DATE(start_date))
  OR
  (DATE(end_date) = DATE_ADD(DATE(start_date), INTERVAL 1 DAY)))
)
SELECT 
  subscriber_type,
  Hour_Flag,
  dayinterval,
  duration_bucket,
  avg(Duration_in_Minutes) as Avg_Duration,
  count(*) as Trip_Count
FROM
  cte
WHERE
  dayinterval in ("Morning", "Afternoon", "Evening")
GROUP BY subscriber_type, Hour_Flag, dayinterval, duration_bucket
HAVING Trip_Count > 2000
  AND subscriber_type = "Customer"
  AND duration_bucket = "Less than 30 min"
ORDER BY 2,1 '  > Potential_Commuters.csv
```

### Stations - High vs. Low Usage

One major aspect is the level of usage across stations.  There are stations where average daily trip counts are 1000+ and those where it is below 10.  Boosting the usage at under-utilized stations is one way to increase ridership.  This an be done through special offers at these highly under-utilized stations.

The query below captures the counts across various stations for each day of the week.  The result is categorized into buckets based on the range in which they fall with respect to total trips in a day. 

```python
%%bash
bq query --use_legacy_sql=FALSE --format=csv --max_rows=2000 '
with cte as
(
SELECT
  #EXTRACT(YEAR FROM start_date) AS Year,
  FORMAT_DATE("%a", DATE(start_date))  AS Weekday_Flag,
  subscriber_type as Subscriber_Type,
  start_station_name as Start_Station,
  count(*) as Count_By_Day,
  round(avg(duration_sec)/60, 2) as Average_Duration_Minutes
FROM
  `bigquery-public-data.san_francisco.bikeshare_trips`
WHERE 
  EXTRACT(YEAR FROM start_date) = 2016
GROUP BY 
  Weekday_Flag, Subscriber_Type, Start_Station 
)
select 
  *
from 
  cte
ORDER BY 1 asc, 4 desc ' > CountByDay_ForHeatMap.csv
```

```python
tripcount = pd.read_csv('CountByDay_ForHeatMap.csv')

# add cuts for ranges (trip counts)
ranges = [0,10,20,50,100,500,1000,1500,2000, 3000]
labels = ['1-10','10-20','20-50','50-100','100-500','500-1000','1000-1500','1500-2000','2000+']
tripcount['TripCounts'] = pd.cut(tripcount.Count_By_Day, ranges, labels=labels).astype('category')
tripcount.head()
```

The heatmap below depicts the average trips at stations in each of the categories.  This is a good indication of the station that generate high vs. revenue.

```python
fig, ax = plt.subplots(figsize=(16,8))
df = tripcount.groupby(['Weekday_Flag', 'TripCounts']).agg({'Count_By_Day':'mean', 'Average_Duration_Minutes':'mean'})
trips = df.pivot_table(index='Weekday_Flag', columns='TripCounts', values='Count_By_Day', aggfunc=np.mean)
sns.heatmap(trips, annot=True, fmt=".1f", cmap="GnBu")
```
